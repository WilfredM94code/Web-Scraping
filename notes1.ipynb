Web Scrapping is a discipline centered in gathering information from the web
in an automated manner.
A web scraping bot or scrapy spyders systematically execute the following process:
1 - HTTP request to servers to gather raw resources
2 - Raw resources get processed until useful information is gathered
3 - Clean information is now available for a porpuseful usage

Data requests are delivered using an HTTP requests to a server. If request
is successfully processed by the server it will give a response to the client.
Data response can be in the form of HTML files, JSON, XML or in any other content.

Note that there is a way to identify online resources and that is through and
URI which is an UNIFORM RESOURCE IDENTIFIER. This URI can be either an Uniform Resource Localizer (or URL) or a Uniform Resource Name (or URN)

HTTP libraries

There is a multitude of HTTP libraries to execute HTTP requests to servers:
* URLLib: Part if Python standard library
* URLLib2: URLLib for Python 2.X
* Requests: Really easy to use
* HTTPlib:  
* HTTPlib2: Google version of HTTPlib. More control over an HTTP request

```python
import webbrowser
from pprint import pprint

# Simple HTTP request from webbrowser
uri_a = 'https://httpbin.org'
webbrowser.open(uri_a)
```

```python
# HTTP requests from httplib2
import httplib2
http_a = httplib2.Http()
print ('http_a = ',http_a)
print ('type(http_a)',type(http_a))
# This object has a set of methods to execute HTTP requests
```

```python
# Http reaponse and data
resp_a, resp_data_a = http_a.request(uri_a)

# Http reaponse
pprint (resp_a)
print ('type(resp_a) = ',type(resp_a),'len(resp_a) = ',len(resp_a))
print ('resp_a.status ,resp_a.reason ,resp_a.version = ',resp_a.status ,resp_a.reason ,resp_a.version)
# This attribtes mean:
# 1 - Status: Status code from the HTTP response
# 2 - Reason: HTTP meaning of the Status code provided by server
# 3 - Version: The version of HTTP that the server supports. A value of 11 means ‘1.1’.
```

```python
# Http data
pprint (resp_data_a)
print ('type(resp_data_a) = ',type(resp_data_a))
```

```python
# To get a string containing the HTML file content there can be used the 
# following method
html_a = resp_data_a.decode('UTF-8')
pprint(html_a)
print(type(html_a))
# The decoding protocol depends on the type of data recieved by the server
```

```python
# Post HTTP request
# HTTP request to send data to a server through and URI for creating a resource
# HTTP post request most provide the following information:
# 1 - Header: Will provide information to the server on to how to deal with the data provided
# 2 - Body: The data formatted in the manner specified on headers

# Data declaration
post_data_a = '{"name": "Jack", "college": "LSD magic mushroom"}'
headers_a =  {'content-type': 'application/json'}
uri_d = 'https://httpbin.org/post'

# Request
resp_g, resp_data_g = http_a.request(uri_d, method = 'POST', body = post_data_a, headers= headers_a)

# Response
pprint(resp_g)
```

```python
# Given the nature of the HTTP connection there can be used a 'with' statement to
# use the connection and close it once a procees is done.

# Note that the next example allows to deal with errors within the code
from urllib.error import URLError, HTTPError

# Error intended
try:
    with urllib.request.urlopen('https://google.com/search?q=python') as resp_data_m:
        pprint (resp_data_m.read().decode('ISO-8859-1'))
except HTTPError as err:
    print (err.reason, err.code)
```

```python
# Beautiful Soup
# Beautiful Soup is a Python library to analyze or parse HTML documents. Beautiful 
# soup structures HTML/XML data as a tree with every tag being a different branch.
# It provides valuable tools to extract info from HTML files which makes it
# useful for web scraping

# This structure considers a hierarchical tag relation. 
# 1 - The HTML tag is considered parentNode, and is considered the parentNode for 
# every other tag.
# 2 - The first nested tag is considered the firstChild and the inmediate nested tag
# from a reference tag is considered the nextSibling. 
# 3 - If there are not nested tags within a tag that tag will be considered the 
# lastChild. Every
# 4 - Every nested tad have a upper level tag considered previousSibling
```

```python
# Beautiful soup
from bs4 import BeautifulSoup

# HTML string
html_b = """

<html>
    <head>
        <title>The Dormouse's story</title>
    </head>
<body>
<p class="title">
    <b>The Dormouse's story</b>
</p>

<p class="story">
Once upon a time there were three little sisters; and their names were:

    <a href="http://example.com/elsie" class="sister" id="link1">Elsie</a>,
    <a href="http://example.com/lacie" class="sister" id="link2">Lacie</a> and
    <a href="http://example.com/tillie" class="sister" id="link3">Tillie</a>;

and they lived at the bottom of a well.
</p>

<p class="story">The story continues</p>
</body>
</html>
"""

# Beautiful soup pasrese data from an HTML string like the one shown above
soup_a = BeautifulSoup(html_b)
print (soup_a)
print (type(soup_a))
```

```python
# Beautiful soup has a method called 'prettify()' that places correct indentation
# on an HTML string
print (soup_a.prettify())
```

```python
# Tags can be accessed by using tag names a attributes
print (soup_a.title)
```

```python
# When parent tags are called bring all content within the open and closing tag
print (soup_a.body)
```

```python
# In case there are several intances of a tag BS will return the 1st one
print (soup_a.a)
```

```python
# There can be called tag names
print (soup_a.body.name)
```

```python
# There can be called parent tag
print (soup_a.a.parent)
```
